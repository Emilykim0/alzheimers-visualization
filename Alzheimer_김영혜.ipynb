### [3] 데이터 분석 및 시각화

1. dataDF에서 혈압 및 콜레스테롤 관련 변수 선택

```python
relevant_columns = ["Hypertension", "SystolicBP", "DiastolicBP", 
                    "CholesterolTotal", "CholesterolLDL", "CholesterolHDL", 
                    "CholesterolTriglycerides","MMSE", "Diagnosis"]
# 관련 데이터만 추출
ad_data = adDF1[relevant_columns]
ad_data.head()
```

```python
ad_data['Diagnosis'].value_counts()
```

```python
# 지표와 진단결과 상관관계 ->heatmap
# dataDF 자리에 본인 지표 부분 DF 대입
# Diagnosis 컬럼이 포함된 DF일 것!!

# 상관관계계수 계산
correlation = ad_data.corr(numeric_only=True)

# 히트맵 모양 대각선으로 자르기
mask = np.triu(np.ones_like(correlation, dtype=bool))
# np.ones_like() 함수로 correlation과 동일한 모양과 크기를 가지는 모든 요소가 True인 배열을 생성
# np.triu() 함수는 상삼각 행렬(Upper triangular matrix)를 생성
# 즉, 대각선 및 그 위의 요소들은 True, 그 아래의 요소들은 False로 설정됨

# 상관관계계수 대입
plt.figure(figsize=(8, 7))
cmap=sns.color_palette("YlOrBr_r", as_cmap=True)
sns.heatmap(correlation,
            cmap=cmap, cbar_kws={"shrink": .5}, # 절대값 사용하면 색상이 진할수록 관계가 있다는 것을 의미함
            mask=mask,annot=correlation, fmt=".2f")  # 히트맵 위에 표시되는 값은 부호가 있는 상관관계계수 값

plt.show()
```

- AD 진단 기준인 MMSE (점수가 낮을 수록 AD 환자)를 제외한 나머지 지표에서 상관관계가 보이지 않음
- 그래도 일단 시각화 분석 해보기
- 다른 관련 변수 필요 -> adDF2, adDF3

```python
# 진단 여부에 따른 혈압 및 콜레스테롤 분포 (Boxplot)
# 그래프 레이아웃 설정 (2행 3열)
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# 색상 코드
colors=utils.color3

# Boxplot을 위한 변수 목록 (혈압 및 콜레스테롤 관련 변수)
bp_vars = ["SystolicBP", "DiastolicBP", "Hypertension",
           "CholesterolTotal", "CholesterolLDL", "CholesterolHDL"]

# 각 Boxplot 생성
for i, var in enumerate(bp_vars):
    # divmod() : 나눗셈을 했을 때 몫과 나머지를 동시에 구하는 함수
    row, col = divmod(i, 3)  # 위치 계산
    axes[row, col].boxplot(
        [adDF1[adDF1['Diagnosis'] == d][var].dropna() for d in adDF1['Diagnosis'].unique()],
        tick_labels=adDF1['Diagnosis'].unique(),
        patch_artist=True,  # 색상 채우기
        boxprops=dict(facecolor=colors[i])  # 개별 색상 지정
    )
    axes[row, col].set_title(f"{var} by AD Diagnosis")
    axes[row, col].set_xlabel("AD Diagnosis")
    axes[row, col].set_ylabel(var)

# 그래프 레이아웃 정리
plt.tight_layout()
plt.show()
```

```python
# 혈압 및 콜레스테롤에 따른 MMSE 분포 확인 (Scatter)
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# 혈압과 MMSE 관계
axes[0, 0].scatter(ad_data["SystolicBP"], ad_data["MMSE"], alpha=0.5,color=utils.color3[0])
axes[0, 0].set_title("SystolicBP vs MMSE")
axes[0, 0].set_xlabel("Systolic Blood Pressure")
axes[0, 0].set_ylabel("MMSE Score")

axes[0, 1].scatter(ad_data["DiastolicBP"], ad_data["MMSE"], alpha=0.5,color=utils.color3[1])
axes[0, 1].set_title("DiastolicBP vs MMSE")
axes[0, 1].set_xlabel("Diastolic Blood Pressure")
axes[0, 1].set_ylabel("MMSE Score")

axes[0, 2].scatter(ad_data["Hypertension"], ad_data["MMSE"], alpha=0.5,color=utils.color3[2])
axes[0, 2].set_title("Hypertension vs MMSE")
axes[0, 2].set_xlabel("Hypertension (0=No, 1=Yes)")
axes[0, 2].set_ylabel("MMSE Score")

# 콜레스테롤과 MMSE 관계
axes[1, 0].scatter(ad_data["CholesterolTotal"], ad_data["MMSE"], alpha=0.5,color=utils.color3[3])
axes[1, 0].set_title("Total Cholesterol vs MMSE")
axes[1, 0].set_xlabel("Total Cholesterol")
axes[1, 0].set_ylabel("MMSE Score")

axes[1, 1].scatter(ad_data["CholesterolLDL"], ad_data["MMSE"], alpha=0.5,color=utils.color3[4])
axes[1, 1].set_title("LDL Cholesterol vs MMSE")
axes[1, 1].set_xlabel("LDL Cholesterol")
axes[1, 1].set_ylabel("MMSE Score")

axes[1, 2].scatter(ad_data["CholesterolHDL"], ad_data["MMSE"], alpha=0.5,color=utils.color3[5])
axes[1, 2].set_title("HDL Cholesterol vs MMSE")
axes[1, 2].set_xlabel("HDL Cholesterol")
axes[1, 2].set_ylabel("MMSE Score")

plt.tight_layout()
plt.show()
```

```python
# AD 진단 여부에 따른 평균 혈압 및 콜레스테롤 값 비교 (Bar Plot)
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# 그룹별 평균 계산
grouped_data = ad_data.groupby("Diagnosis").mean()

# 색상 코드
colors=utils.color2[1:4:2]
barwidth=0.4

# 혈압 평균 비교
axes[0, 0].bar(["Non-AD", "AD"], grouped_data["SystolicBP"], color=colors, width=barwidth)
axes[0, 0].set_title("Average Systolic Blood Pressure by AD Diagnosis")
axes[0, 0].set_ylabel("Systolic Blood Pressure")

axes[0, 1].bar(["Non-AD", "AD"], grouped_data["DiastolicBP"], color=colors, width=barwidth)
axes[0, 1].set_title("Average Diastolic Blood Pressure by AD Diagnosis")
axes[0, 1].set_ylabel("Diastolic Blood Pressure")

axes[0, 2].bar(["Non-AD", "AD"], grouped_data["Hypertension"], color=colors,width=barwidth)
axes[0, 2].set_title("Hypertension Prevalence by AD Diagnosis")
axes[0, 2].set_ylabel("Proportion with Hypertension")

# 콜레스테롤 평균 비교
axes[1, 0].bar(["Non-AD", "AD"], grouped_data["CholesterolTotal"], color=colors,width=barwidth)
axes[1, 0].set_title("Average Total Cholesterol by AD Diagnosis")
axes[1, 0].set_ylabel("Total Cholesterol")

axes[1, 1].bar(["Non-AD", "AD"], grouped_data["CholesterolLDL"], color=colors, width=barwidth)
axes[1, 1].set_title("Average LDL Cholesterol by AD Diagnosis")
axes[1, 1].set_ylabel("LDL Cholesterol")

axes[1, 2].bar(["Non-AD", "AD"], grouped_data["CholesterolHDL"], color=colors, width=barwidth)
axes[1, 2].set_title("Average HDL Cholesterol by AD Diagnosis")
axes[1, 2].set_ylabel("HDL Cholesterol")

plt.tight_layout()
plt.show()
```

```python
# 혈압 및 콜레스테롤 각 구간별 AD 진단 환자 비율 (Bar Plot) 

fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# 변수 리스트
variables = ["SystolicBP", "DiastolicBP", "Hypertension", "CholesterolTotal", "CholesterolLDL", "CholesterolHDL"]

# 각 변수에 대해 그룹별 AD 진단률 계산 및 시각화
for i, var in enumerate(variables):
    row, col = divmod(i, 3)  # 2행 3열 subplot 위치 계산

    # 적절한 구간 크기 설정 (혈압은 10단위, 콜레스테롤은 20-50단위)
    if "BP" in var:  # 혈압 관련 변수
        bin_size = 10
    elif "Cholesterol" in var:  # 콜레스테롤 관련 변수
        bin_size = 20
    else:
        bin_size = 1  # Hypertension과 같은 이진 변수는 구간화 필요 없음

    # 변수 구간화
    if bin_size > 1:
        ad_data=ad_data.copy()
        bins = range(int(ad_data[var].min()), int(ad_data[var].max()) + bin_size, bin_size)
        ad_data[f"{var}_binned"] = pd.cut(ad_data[var], bins=bins, right=False)
        grouped = ad_data.groupby(f"{var}_binned", observed=False)["Diagnosis"].mean()
    else:
        grouped = ad_data.groupby(var, observed=False)["Diagnosis"].mean()  # 이진 변수는 그대로

    axes[row, col].bar(grouped.index.astype(str), grouped.values, color=utils.color3)
    axes[row, col].set_title(f"AD Diagnosis Rate by {var}")
    axes[row, col].set_xlabel(var)
    axes[row, col].set_ylabel("AD Diagnosis Rate")
    axes[row, col].tick_params(axis='x', rotation=45) 

plt.tight_layout()
plt.show()
```

```python
# ETIV, NWBV, And ASF By CDR Score & MMSE
# 각 지표별 적절한 구간 설정 (10, 20, 30 등의 가독성 좋은 방식으로 구간화)
bin_settings = {
    "eTIV": [1000, 1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900, 2000],
    "nWBV": [0.60, 0.65, 0.70, 0.75, 0.80],
    "ASF": [0.8, 0.9, 1.0, 1.1, 1.2, 1.3],
}

# 각 지표별 subplot을 위한 figure 설정
fig, axes = plt.subplots(3, 2, figsize=(14, 12))

# 지표 리스트
metrics = ["eTIV", "nWBV", "ASF"]
titles = ["MMSE Score by eTIV Range", "CDR Score by eTIV Range",
          "MMSE Score by nWBV Range", "CDR Score by nWBV Range",
          "MMSE Score by ASF Range", "CDR Score by ASF Range"]

# 구간 나누기 함수 (연속형 변수)
def bin_column(data, bins):
    return pd.cut(data, bins=bins, include_lowest=True, duplicates="drop")

# 각 지표별로 subplot 생성
for i, metric in enumerate(metrics):
    ad_data33=ad_data33.copy()
    # 구간화
    ad_data33[f"{metric}_bin"] = bin_column(ad_data33[metric], bin_settings[metric])
    
    # MMSE 점수 바 그래프 (X축: 지표 구간, Y축: MMSE 평균)
    mmse_bins = ad_data33.groupby(f"{metric}_bin",observed=False)["MMSE"].mean().dropna()
    mmse_bins.plot(kind="bar", ax=axes[i, 0], color=utils.color3)
    axes[i, 0].set_title(titles[i * 2], fontsize=14)
    axes[i, 0].set_ylabel("Mean MMSE Score")
    axes[i, 0].set_xlabel(metric)
    axes[i, 0].tick_params(axis='x', rotation=45)

    # CDR 점수 바 그래프 (X축: 지표 구간, Y축: CDR 평균)
    cdr_bins = ad_data33.groupby(f"{metric}_bin",observed=False)["CDR"].mean().dropna()
    cdr_bins.plot(kind="bar", ax=axes[i, 1], color=utils.color3)
    axes[i, 1].set_title(titles[i * 2 + 1], fontsize=14)
    axes[i, 1].set_ylabel("Mean CDR Score")
    axes[i, 1].set_xlabel(metric)
    axes[i, 1].tick_params(axis='x', rotation=45)

# 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
# 데이터 준비: APOE4 발현 케이스 분류
def classify_apoe4_case(row):
    if row["APOE.a1"] != 4 and row["APOE.a2"] != 4:
        return "None"
    elif row["APOE.a1"] == 4 and row["APOE.a2"] != 4:
        return "A1 Only"
    elif row["APOE.a1"] != 4 and row["APOE.a2"] == 4:
        return "A2 Only"
    elif row["APOE.a1"] == 4 and row["APOE.a2"] == 4:
        return "Both"

ad_data2["APOE4_Case"] = ad_data2.apply(classify_apoe4_case, axis=1)
```
